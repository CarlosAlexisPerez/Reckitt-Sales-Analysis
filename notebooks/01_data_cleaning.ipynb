```python
# Cargar las tablas de datos en Pandas:

# Utiliza la función de Pandas para cargar cada tabla de datos desde el archivo csv o xlsx (DIM_CATEGORY, DIM_PRODUCT, 
# DIM_SEGMENT, DIM_CALENDAR, FACT_SALES) en un DataFrame.

import pandas as pd

DIM_CATEGORY = pd.read_csv('DIM_CATEGORY (2).csv')
DIM_PRODUCT = pd.read_excel('DIM_PRODUCT (1).xlsx')
DIM_SEGMENT = pd.read_excel('DIM_SEGMENT (1).xlsx')
DIM_CALENDAR = pd.read_excel('DIM_CALENDAR (2).xlsx')
FACT_SALES = pd.read_csv('FACT_SALES (1).csv')
```

```python
# 2 REVISIÓN DE LOS DATOS

# Realiza una revisión rápida de las primeras filas de cada DataFrame para entender la estructura de los datos y sus principales 
# características. Utiliza métodos como head(), info() y describe() para tener una idea general del contenido de cada tabla.

# DIM_CATEGORY

print('\nDIM_CATEGORY')
print('\nInicio')     
print(DIM_CATEGORY.head())
print('\nInformación') 
print(DIM_CATEGORY.info())
print('\nDescipción') 
print(DIM_CATEGORY.describe(include = 'all'))
```

```python
# DIM_PRODUCT

print('\nDIM_PRODUCT')
print('\nInicio')     
print(DIM_PRODUCT.head())
print('\nInformación') 
print(DIM_PRODUCT.info())
print('\nDescipción') 
print(DIM_PRODUCT.describe(include = 'all'))
```

```python
# DIM_SEGMENT

print('\nDIM_SEGMENT')
print('\nInicio')     
print(DIM_SEGMENT.head())
print('\nInformación') 
print(DIM_SEGMENT.info())
print('\nDescipción') 
print(DIM_SEGMENT.describe(include = 'all'))
```

```python
# DIM_CALENDAR

print('\nDIM_CALENDAR')
print('\nInicio')     
print(DIM_CALENDAR.head())
print('\nInformación') 
print(DIM_CALENDAR.info())
print('\nDescipción') 
print(DIM_CALENDAR.describe(include = 'all'))
```

```python
# FACT_SALES

print('\nFACT_SALES')
print('\nInicio')     
print(FACT_SALES.head())
print('\nInformación') 
print(FACT_SALES.info())
print('\nDescipción') 
print(FACT_SALES.describe(include = 'all'))
```

```python
# 3 LIMPIEZA DE DATOS

# Normalizar nombres de columnas
for df in [DIM_CATEGORY, DIM_PRODUCT, DIM_SEGMENT, DIM_CALENDAR, FACT_SALES]:
    df.columns = (df.columns.str.strip().str.lower().str.replace(' ', '_'))

# Limpieza de texto
def limpiar_texto(df):
    for col in df.select_dtypes(include='object').columns:
        df[col] = df[col].astype(str).str.strip()
    return df

DIM_CATEGORY = limpiar_texto(DIM_CATEGORY)
DIM_PRODUCT = limpiar_texto(DIM_PRODUCT)
DIM_SEGMENT = limpiar_texto(DIM_SEGMENT)
DIM_CALENDAR = limpiar_texto(DIM_CALENDAR)
FACT_SALES = limpiar_texto(FACT_SALES)

print('\nCOLUMNAS LIMPIAS:')
print(DIM_CATEGORY.columns)
print(DIM_PRODUCT.columns)
print(DIM_SEGMENT.columns)
print(DIM_CALENDAR.columns)
print(FACT_SALES.columns)

print('\nEJEMPLO TEXTO LIMPIO (DIM_CATEGORY):')
print(DIM_CATEGORY.head())
```

```python
# Detectar valores Nulos

def limpiar_nulos(df):
    for col in df.columns:
        if df[col].dtype == 'object':
            df[col] = df[col].fillna('Desconocido')
        else:
            df[col] = df[col].fillna(df[col].median())
    return df

DIM_CATEGORY = limpiar_nulos(DIM_CATEGORY)
DIM_PRODUCT = limpiar_nulos(DIM_PRODUCT)
DIM_SEGMENT = limpiar_nulos(DIM_SEGMENT)
DIM_CALENDAR = limpiar_nulos(DIM_CALENDAR)
FACT_SALES = limpiar_nulos(FACT_SALES)

print('\nVERIFICACIÓN DE VALORES NULOS:\n')

print('DIM_CATEGORY')
print(DIM_CATEGORY.isnull().sum())

print('\nDIM_PRODUCT')
print(DIM_PRODUCT.isnull().sum())

print('\nDIM_SEGMENT')
print(DIM_SEGMENT.isnull().sum())

print('\nDIM_CALENDAR')
print(DIM_CALENDAR.isnull().sum())

print('\nFACT_SALES')
print(FACT_SALES.isnull().sum())

print('\nMUESTRA DIM_CATEGORY')
print(DIM_CATEGORY.head())

print('\nMUESTRA FACT_SALES')
print(FACT_SALES.head())
```

```python
# Eliminación de Duplicados

DIM_CATEGORY.drop_duplicates(inplace=True)
DIM_PRODUCT.drop_duplicates(inplace=True)
DIM_SEGMENT.drop_duplicates(inplace=True)
DIM_CALENDAR.drop_duplicates(inplace=True)
FACT_SALES.drop_duplicates(inplace=True)

print('\nVERIFICACIÓN DE DUPLICADOS:\n')

print('DIM_CATEGORY duplicados:', DIM_CATEGORY.duplicated().sum())
print('DIM_PRODUCT duplicados:', DIM_PRODUCT.duplicated().sum())
print('DIM_SEGMENT duplicados:', DIM_SEGMENT.duplicated().sum())
print('DIM_CALENDAR duplicados:', DIM_CALENDAR.duplicated().sum())
print('FACT_SALES duplicados:', FACT_SALES.duplicated().sum())

print('\nTAMAÑO DE LAS TABLAS:\n')

print('DIM_CATEGORY:', DIM_CATEGORY.shape)
print('DIM_PRODUCT:', DIM_PRODUCT.shape)
print('DIM_SEGMENT:', DIM_SEGMENT.shape)
print('DIM_CALENDAR:', DIM_CALENDAR.shape)
print('FACT_SALES:', FACT_SALES.shape)
```

```python
# 4 Unión de DataFrames

# Ventas + Productos
ventas = FACT_SALES.merge(DIM_PRODUCT, left_on = 'item_code', right_on = 'item', how = 'left')

# + Segmentos
ventas = ventas.merge(DIM_SEGMENT, on = ['category', 'attr1', 'attr2', 'attr3', 'format'], how = 'left')

# + Categorías
ventas = ventas.merge(DIM_CATEGORY, left_on = 'category', right_on = 'id_category', how = 'left')

# + Calendario
ventas = ventas.merge(DIM_CALENDAR, on = 'week', how = 'left')

print(ventas[['item_code', 'item', 'segment', 'category_y', 'date']].head())

# Limpieza mínima post-unión (evitar confusión de columnas)
ventas.rename(columns={
    'category_y': 'category_name',
    'segment': 'segment_name'
}, inplace=True)

ventas.drop(columns=['category_x', 'id_category'], inplace=True, errors='ignore')

# Verificación rápida
print(ventas[['item_code', 'item', 'segment_name', 'category_name', 'date']].head())
```

```python
# 5 Aplicar Transformaciones 

# Estandarización de formatos

# Convertir columnas de fecha a formato datetime
if 'date' in ventas.columns:
    ventas['date'] = pd.to_datetime(ventas['date'], errors='coerce')

if 'year' in ventas.columns:
    ventas['year'] = ventas['year'].astype(int)

if 'month' in ventas.columns:
    ventas['month'] = ventas['month'].astype(int)

# Estandarizar texto categórico
columnas_texto = ventas.select_dtypes(include ='object').columns

for col in columnas_texto:
    ventas[col] = ventas[col].str.upper()

# Creación de columnas calculadas

# Total de venta
if 'total_unit_sales' in ventas.columns and 'total_value_sales' in ventas.columns:
    ventas['total_venta'] = ventas['total_value_sales']
else:
    raise KeyError("No existen columnas para calcular total_venta")

# Año - Mes
ventas['year_month'] = ventas['date'].dt.to_period('M').astype(str)

# Normalización de nombres finales
if 'category_y' in ventas.columns:
    ventas.rename(columns={'category_y': 'category_name'}, inplace=True)
else:
    ventas['category_name'] = 'NO DEFINIDO'

if 'segment' in ventas.columns:
    ventas.rename(columns={'segment': 'segment_name'}, inplace=True)
else:
    ventas['segment_name'] = 'NO DEFINIDO'
    
# Agrupaciones para análisis

# Ventas por categoría
ventas_categoria = ventas.groupby('category_name').agg(total_ventas = ('total_venta', 'sum'),
    unidades_vendidas=('total_unit_sales', 'sum')).reset_index()

# Ventas por segmento
ventas_segmento = ventas.groupby('segment_name').agg(
    total_ventas = ('total_venta', 'sum')).reset_index()

# Ventas mensuales
ventas_mensuales = ventas.groupby('year_month').agg(
    total_ventas = ('total_venta', 'sum')).reset_index()

# Verificación rápida de transformaciones
print(ventas[['total_unit_sales', 'total_value_sales', 'total_venta', 'year_month']].head())

# Resultado Final

print("DATAFRAME FINAL")
print(ventas.head())
print(ventas.info())
```

```python
# Guardar DataFrame consolidado en CSV

ventas.to_csv('ventas_consolidadas.csv', index=False, encoding='utf-8-sig')

print("Archivo Proyecto_empresa_aliada: entregable 1.csv guardado correctamente")
```
